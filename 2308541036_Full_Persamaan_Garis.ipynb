# Persamaan Garis yang Melewati 2 Titik

a. Berpotongan

```python
# Persamaan garis yang melewati 2 titik
# Garis pertama melewati (i,j) dan (k,l)
# Garis kedua melewati (m,n) dan (o,p)
# Terdapat 3 kemungkinan dari kedua persamaan garis ini:
# 1. Terdapat perpotongan di satu titik (gradien kedua garis tidak sama)
# 2. Terdapat tak berhingga banyaknya solusi (kedua garis berhimpitan) (kedua garis sejajar)
# 3. Tidak terdapat perpotongan garis (tidak terdapat solusi) (kedua garis sejajar)

import numpy as np
import matplotlib.pyplot as plt

print('===============Selamat datang di pemrograman mencari solusi dari bentuk persamaan garis yang melewati 2 titik===============')
#persamaan garis 1 melewati titik (i,j) dan (k,l)
i_36 = float(input("Masukkan nilai x dari titik pertama dari garis pertama adalah "))
j_36 = float(input("Masukkan nilai y dari titik pertama dari garis pertama adalah "))
k_36 = float(input("Masukkan nilai x dari titik kedua dari garis pertama adalah "))
l_36 = float(input("Masukkan nilai y dari titik kedua dari garis pertama adalah "))
#persamaan garis 2 melewati titik (m,n) dan (o,p)
m_36 = float(input("Masukkan nilai x dari titik pertama dari garis kedua adalah "))
n_36 = float(input("Masukkan nilai y dari titik pertama dari garis kedua adalah "))
o_36 = float(input("Masukkan nilai x dari titik kedua dari garis kedua adalah "))
p_36 = float(input("Masukkan nilai y dari titik kedua dari garis kedua adalah "))

#hitung gradien persamaan garis 1 dan 2
m_1_36 = (l_36-j_36)/(k_36-i_36) # Gradien garis pertama
m_2_36 = (p_36-n_36)/(o_36-m_36) # Gradien garis kedua

# Persamaan garis pertama adalah y=m_1(x-i)+j=m_1x-m_1i+j
# Persamaan garis kedua adalah y=m_2(x-m)+n=m_2x-m_2m+n

#hitung titik potong garis pertama dan garis kedua terhadap sumbu y
c_1_36 = -m_1_36*i_36+j_36 # Titik potong garis pertama terhadap sumbu y
c_2_36 = -m_2_36*m_36+n_36 # titik potong garis kedua terhadap sumbu y garis kedua

print("Ini adalah persamaan garis pertamanya y={}x+{}".format(m_1_36,c_1_36))
print("Ini adalah persamaan garis keduanya y={}x+{}".format(m_2_36,c_2_36))

if m_1_36==m_2_36:
    #print("Kedua garis sejajar sehingga kemungkinannya adalah tidak terdapat solusi atau terdapat tak berhingga banyaknya solusi")
    if c_1_36 == c_2_36:
        print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
        x=np.linspace(-5,5,100)
        plt.plot(x,m_1_36*(x-i_36)+j_36,color='brown')
        plt.plot(x,m_2_36*(x-m_36)+n_36,color='green')
        plt.show()
    else:
        print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
        x=np.linspace(-5,5,100)
        plt.plot(x,m_1_36*(x-i_36)+j_36,color='brown')
        plt.plot(x,m_2_36*(x-m_36)+n_36,color='green')
        plt.show()
else:
    print("Akan terdapat satu titik potong")
    tp_x_36 = (-m_2_36*m_36+n_36+m_1_36*i_36-j_36)/(m_1_36-m_2_36)
    tp_y_36 = m_1_36*(tp_x_36-i_36)+j_36
    print("Titik potong kedua garis adalah ({},{})".format(tp_x_36,tp_y_36))

    x=np.linspace(tp_x_36-5,tp_x_36+5,100)
    plt.plot(x,m_1_36*(x-i_36)+j_36,color='brown')
    plt.plot(x,m_2_36*(x-m_36)+n_36,color='green')
    plt.axvline(x=0,color='grey')
    plt.axhline(y=0,color='grey')
    plt.show()

print('==========================================Terima Kasih==========================================')
```

# Persamaan Garis jika Diketahui Gradien dan Titik yang Dilewati

```python
# Memanggil numpy
import numpy as np
# Memanggil matplotlib.pyplot
import matplotlib.pyplot as plt

print('===============Selamat datang di pemrograman mencari solusi dari bentuk persamaan garis jika diketahui gradien dan titik yang dilewati===============')
print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
# Masukkan nilai gradien persamaan garis pertama dan titik yang diketahui (i,j)
m1_36 = float(input("Masukkan gradien dari garis pertama: "))
i_36 = float(input("Msukkan nilai x dari titik yang dilewati oleh garis pertama: "))
j_36 = float(input("Masukkan nilai y dari titik yang dilewati oleh garis pertama: "))

# Masukkan nilai gradien persamaan garis pertama dan titik yang diketahui (k,l)
m2_36 = float(input("Masukkan gradien dari garis kedua: "))
k_36 = float(input("Masukkan nilai x dari titik yang dilewati oleh garis kedua: "))
l_36 = float(input("Masukkan nilai y dari titik yang dilewati oleh garis kedua: "))

# Menghitung konstanta dari masing-masing persamaan garis
c1_36 = -m1_36*i_36+j_36
c2_36 = -m2_36*k_36+l_36

    # Memeriksa apakah kedua garis sejajar atau tidak
if m1_36 == m2_36:
      # print("Kedua garis sejajar ")
        if c1_36 == c2_36:
            print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
        else:
            print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")

        # Plotting the lines
        x = np.linspace(-10, 10, 400)
        y1_36 = m1_36*(x-i_36)+ j_36
        y2_36 = m2_36*(x-k_36) + l_36
        plt.plot(x, y1_36, label="Garis 1 (y={:.3f}(x-{:.3f})+{:.3f})".format(m1_36, i_36, j_36), color='brown')
        plt.plot(x, y2_36, label="Garis 2 (y={:.3f}(x-{:.3f})+{:.3f})".format(m2_36, k_36, l_36), color='green')
        plt.axvline(x=0, color='black')
        plt.axhline(y=0, color='black')
        plt.legend(loc='best')
        plt.show()
else:
        # Menghitung titik potong dari kedua garis
        tp_x_36 = (c2_36 - c1_36) / (m1_36 - m2_36)
        tp_y_36 = m1_36 * (tp_x_36 - i_36) + j_36
        print("Titik potong kedua garis adalah ({:.3f},{:.3f})".format(tp_x_36, tp_y_36))

        # Plotting the lines
        x = np.linspace(tp_x_36-10, tp_x_36+10, 400)
        y1_36 = m1_36*(x-i_36)+j_36
        y2_36 = m2_36*(x-k_36)+l_36
        plt.plot(x, y1_36, label="Garis 1 (y={:.3f}(x-{:.3f})+{:.3f})".format(m1_36, i_36, j_36), color='brown')
        plt.plot(x, y2_36, label="Garis 2 (y={:.3f}(x-{:.3f})+{:.3f})".format(m2_36, k_36, l_36), color='green')
        plt.scatter(tp_x_36, tp_y_36, color='red', zorder=5)
        plt.axvline(x=0, color='black')
        plt.axhline(y=0, color='black')
        plt.legend(loc='best')
        plt.show()
print('==========================================Terima Kasih==========================================')
```
