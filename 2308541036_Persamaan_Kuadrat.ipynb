# **====PERSAMAAN KUADRAT====**

```python
# Persamaan Kuadrat dengan ax^2+bx+c = 0
# suatu persamaan dapat dicari akar-akarnya dengan cara faktorisasi, maupun rumus abc
# Dalam rumus abc terdapat diskriminan yang dapat dicari dengan : D=b^2-4ac
# Terdapat 3 kemungkinan akar-akar yang akan terbentuk berdasarkan nilai diskriminan
# 1. Jika D < 0;akar imajiner
# 2. Jika D = 0;akar kembar
# 3. Jika D > 0;akar real berbeda
# Berdasarkan nilai a dari suatu persamaan kuadrat ax^2+bx+c=0 dapat terlihat kurva dari grafiknya
# 1. Jika a > 0;kurva membuka keatas
# 2. Jika a < 0;kurva membuka kebawah
# Apabila yang diketahui akar-akar dari suatu persamaan kuadrat, maka dapat dicari persamaan kuadratnya denganÂ (x^2-(a-b)x+ab)
```

```python
# Project 3 (Grafik pada persamaan kuadrat)
# Langkah-langkahnya adalah :
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Memanggil cmath
import cmath
print("--------------------Selamat datang di program Diva untuk menentukan grafik pada persamaan kuadrat--------------------")
print("Yuk, mulai!")
# 4. Input nilai a, b dan c dalam persamaan ax**2 + bx + c = 0
print("Silahkan masukkan nilai konstanta untuk a,b,c")
a_36 = float(input("Nilai konstanta a adalah: "))
b_36 = float(input("Nilai konstanta b adalah: "))
c_36 = float(input("Nilai konstanta c adalah: "))

if a_36 == 0:
    print("Error, nilai yang kamu input bukan persamaan kuadrat")
else:
    print("Nilai yang kamu input dapat membentuk persamaan kuadrat")
    print("Persamaan Kuadratnya adalah ({}x**2 + {}x + {})".format(a_36, b_36, c_36))

    # 5. Input rumus diskriminan
    D_36 = b_36**2 - 4*a_36*c_36
    if D_36 == 0:
        print("Waww, Akar-akarnya kembar")
    elif D_36 > 0:
        print("Akar real berbeda")
    else:
        print("Akarnya imajiner")

    # 6. Input rumus x1, x2
    x1_36 = (-b_36 + cmath.sqrt(D_36)) / (2*a_36)
    x2_36 = (-b_36 - cmath.sqrt(D_36)) / (2*a_36)
    print("Akar-akarnya adalah ({:.2f}, {:.2f})".format(x1_36.real, x2_36.real))

    # 7. Format garis
    x_36 = np.linspace(-10, 10, 400)
    y_36 = a_36*x_36**2 + b_36*x_36 + c_36
    plt.figure(figsize=(8, 6))
    plt.plot(x_36, y_36, label=f'{a_36}x^2 + {b_36}x + {c_36} = 0')

    if D_36 >= 0:
        plt.scatter([x1_36.real, x2_36.real], [0, 0], color='red')
        plt.text(x1_36.real, 0.5, f'({x1_36.real:.2f}, 0)', fontsize=12, ha='center')
        plt.text(x2_36.real, 0.5, f'({x2_36.real:.2f}, 0)', fontsize=12, ha='center')
    else:
        plt.scatter([x1_36.real, x2_36.real], [x1_36.imag, x2_36.imag], color='red')
        plt.text(x1_36.real, x1_36.imag, f'({x1_36.real:.2f}, {x1_36.imag:.2f})', fontsize=12, ha='center')
        plt.text(x2_36.real, x2_36.imag, f'({x2_36.real:.2f}, {x2_36.imag:.2f})', fontsize=12, ha='center')

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x_36')
    plt.ylabel('y_36')
    plt.legend()
    plt.show()
print("----------Terima kasih----------")
```

# 3.1 MENENTUKAN TITIK PUNCAK DAN SUMBU SIMETRI

Contoh 1

```python
# 3.1 Menentukan Titik Puncak dan Sumbu Simetri
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Memanggil cmath
import cmath
print("--------------------Selamat datang di program Diva untuk Menentukan Titik Puncak dan Sumbu Simetri--------------------")
print("Yuk, mulai!")
# 4. Input nilai a,b,c
print("Silahkan masukkan nilai a,b,c")
a_36 = float(input("Nilai a adalah: "))
b_36 = float(input("Nilai b adalah: "))
c_36 = float(input("Nilai c adalah: "))

# 5. Menghitung titik puncak dan sumbu simetri
def hitung_titik_puncak_dan_sumbu_simetri(a_36, b_36, c_36):
    # Sumbu simetri (x) adalah -b / (2a)
    sumbu_simetri_36 = -b_36 / (2 * a_36)

    # Titik puncak (x, y)
    # x dari titik puncak adalah sumbu simetri
    # y dari titik puncak adalah nilai fungsi kuadrat di x = sumbu_simetri
    titik_puncak_x_36 = sumbu_simetri_36
    titik_puncak_y_36 = a_36 * (titik_puncak_x_36 ** 2) + b_36 * titik_puncak_x_36 + c_36

    return (titik_puncak_x_36, titik_puncak_y_36), sumbu_simetri_36

# 6. Menentukan grafik persamaan kuadrat
def plot_persamaan_kuadrat(a_36, b_36, c_36):
    # Menghitung titik puncak dan sumbu simetri
    titik_puncak, sumbu_simetri = hitung_titik_puncak_dan_sumbu_simetri(a_36, b_36, c_36)

    # Membuat rentang x untuk plot
    x_36 = np.linspace(titik_puncak[0] - 10, titik_puncak[0] + 10, 400)
    y_36 = a_36 * x_36**2 + b_36 * x_36 + c_36

    # Plot grafik persamaan kuadrat
    plt.figure(figsize=(10, 6))
    plt.plot(x_36, y_36, label=f'{a_36}x^2 + {b_36}x + {c_36}')

    # Plot titik puncak
    plt.plot(titik_puncak[0], titik_puncak[1], 'ro')  # Titik puncak berwarna merah
    plt.annotate(f'Titik puncak\n({titik_puncak[0]:.2f}, {titik_puncak[1]:.2f})',
                 xy=titik_puncak,
                 xytext=(titik_puncak[0] + 1, titik_puncak[1] + 10),
                 arrowprops=dict(facecolor='black', shrink=0.05))

    # Plot sumbu simetri
    plt.axvline(x=sumbu_simetri, color='gray', linestyle='--', label='Sumbu simetri')

    # Menambahkan label dan judul
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Titik Puncak dan Sumbu Simetri Grafik Persamaan Kuadrat')
    plt.axvline(x=0,color='black')
    plt.axhline(y=0,color='black')
    plt.legend()
    plt.grid(True)
    plt.show()

# 7. Menjalankan fungsi plot_persamaan_kuadrat dengan nilai a, b, dan c yang diberikan
plot_persamaan_kuadrat(a_36, b_36, c_36)
print("----------Terima kasih----------")
```

```python
# Project 4 (Membentuk persamaan kuadrat jika diketahui akar-akarnya)
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Memanggil cmath
import cmath
print("--------------------Selamat datang di program Diva untuk Membentuk persamaan kuadrat jika diketahui akar-akarnya--------------------")
print("Yuk, mulai!")
# 4. Input nilai a, x1 dan x2
print("Silahkan masukkan nilai a, x1 dan x2 nya")
a_36 = float(input("Berapakah nilai a dari persamaan yang akan dibentuk? (tidak boleh sama dengan nol) "))
while a_36 == 0:
    print("Maaf error, nilai a tidak boleh sama dengan nol.")
    a_36 = float(input("Silahkan masukkan nilai a yang tidak sama dengan nol: "))
x1_36 = float(input("Berapakah akar 1 atau x1 dari persamaan yang akan dibentuk? "))
x2_36 = float(input("Berapakah akar 2 atau x2 dari persamaan yang akan dibentuk? "))

# 5. Bentuk persamaan
def bentuk_persamaan_kuadrat(a_36, x1_36, x2_36):
    b_36 = -(x1_36 + x2_36) * a_36
    c_36 = x1_36 * x2_36 * a_36
    persamaan = f"{a_36}x**2 + ({b_36})x + ({c_36}) = 0"
    return persamaan

# 6. Menampilkan persamaan
persamaan = bentuk_persamaan_kuadrat(a_36, x1_36, x2_36)
print("Berikut merupakan bentuk persamaan kuadratnya adalah " + persamaan)
print("----------Terima kasih----------")
```
