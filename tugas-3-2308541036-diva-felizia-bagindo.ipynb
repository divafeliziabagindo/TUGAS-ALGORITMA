```python
#Menghitung total penjumlahan angka genap dari 1 sampai 20
total = 0
bilangan = 1
while bilangan <= 20:
    if bilangan % 2 == 0:
        total += bilangan
    bilangan += 1
print("Total penjumlahan angka genap dari 1 sampai 20 adalah sebanyak:", total)
```

```python
# Perintah untuk memasukkan batas atas deret Fibonacci
batas_atas = int(input("Masukkan batas atas deret Fibonacci: "))

# Inisialisasi dua angka pertama dalam deret Fibonacci
angka_sebelumnya = 0
angka_sekarang = 1

# Menampilkan dua angka pertama
print("Deret Fibonacci:")
print(angka_sebelumnya)
print(angka_sekarang)

# Menghitung dan menampilkan deret Fibonacci menggunakan loop while
while angka_sekarang + angka_sebelumnya <= batas_atas:
    angka_selanjutnya = angka_sekarang + angka_sebelumnya
    print(angka_selanjutnya)
    angka_sebelumnya, angka_sekarang = angka_sekarang, angka_selanjutnya
```
