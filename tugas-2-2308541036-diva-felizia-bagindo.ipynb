```python
#Formula for the Volume of a Cylinder
pi = 3.14
r = float(input("Input the radius of a circle: "))
base_area = pi*(r**2)
h = float(input("Input the height of cylinder: "))
volume = base_area*h
print("The Volume of a Cylinder is " + str(volume))
print("You did it!")
```

```python
x = float(input("Masukkan nilai dari x: "))
y = float(input("Masukkan nilai dari y: "))
```

```python
a = x<y
print('x < y : ' + str(a) + ' atau ' + str(~a))
a = x>y
print('x > y : ' + str(a) + ' atau ' + str(~a))
a = x >= y
print('x >= y : ' + str(a) + ' atau ' + str(~a))
a = x <= y
print('x <= y : ' + str(a) + ' atau ' + str(~a))
a = x == y
print('x == y : ' + str(a) + ' atau ' + str(~a))
a = x != y
print('x != y : ' + str(a) + ' atau ' + str(~a))
```

```python
kalimat_c = 'Hari ini cuacanya sangat mendung.'
kalimat_d = 'Cuaca hari ini cerah dan berawan.'
a = kalimat_c < kalimat_d
print('Urutan alfabetic : ' + str(a) + ' atau ' + str(~a))
a = kalimat_c > kalimat_d
print('Urutan alfabetic : ' + str(a) + ' atau ' + str(~a))
a = kalimat_c == kalimat_d
print('Panjang kalimat pertama sama dengan panjang kalimat kedua : ' + str(a) + ' atau ' + str(~a))
a = kalimat_c != kalimat_d
print('Bunyi kalimat pertama tidak sama dengan kalimat kedua : ' + str(a) + ' atau '  + str(~a))
```
