Contoh 1

```python
# 1. Import Library
import numpy as np
import matplotlib.pyplot as plt

print('===============Selamat datang di pemrograman mencari solusi dari bentuk persamaan garis===============')
# 2. Menentukan Titik dan Menghitung Gradien Garis Pertama
# Persamaan garis pertama melewati titik (e,f) dan (g,h)
e_36=1 #Titik (e,f)
f_36=1
g_36=2 #Titik (g,h)
h_36=2
m1_36=(h_36-f_36)/(g_36-e_36) # Menghitung gradien garis pertama
print ("Ini adalah gradien garis pertamanya {}".format(m1_36)) #Mencetak gradien garis pertama
print("Berikut merupakan persamaan garis pertama adalah y={}x+{}".format(m1_36, -m1_36*e_36+f_36)) #Mencetak persamaan garis pertama
#y1 = m1*(x-e)+f

# 3. Menghitung Titik dan Menghitung Gradien Garis Kedua
# Persamaan garis pertama kedua melewati titik (i,j) dan (k,l)
i_36=1 #Titik（i,j）
j_36=1
k_36=2 #Titik（k,l）
l_36=-2
m2_36=(l_36-j_36)/(k_36-i_36) #Menghitung gradien garis kedua
print( "Ini adalah gradien garis ke duanya {}".format(m2_36)) #Mencetak gradien garis kedua
print("Berikut merupakan persamaan garis kedua adalah y={}x+{}".format(m2_36, -m2_36*i_36+j_36)) #Mencetak persamaan garis kedua
#y2=m2*(x-i)+j

# 4. Menentukan Titik Potong Kedua Garis
if m1_36==m2_36:
    print( "Wah gradiennya sama tuh, jadi kemungkinan persamaannya tidak ada solusi atau solusinya ada banyak.") #Jika gradien kedua garis sama, maka garis-garis tersebut sejajar dan tidak memiliki titik potong atau memiliki banyak solusi (jika garisnya sama)
else:
    # Titik potong kedua garis pada (tp_x, tp_y)
    tp_x_36 = (m1_36*e_36-f_36-m2_36*i_36+j_36)/(m1_36-m2_36)
    tp_y_36 = m2_36*(tp_x_36-i_36)+j_36

    print ("Titik potong untuk x dan y adalah ({},{})".format(tp_x_36,tp_y_36)) #Mencetak titik potong kedua garis

# 5. Plotting Garis dan Titik Potong
    x_36=np.arange(tp_x_36-5,tp_x_36+5,0.01) #Membuat array x untuk plotting
    plt.figure(figsize=(4,4),dpi=150) #Menentukan ukuran dan resolusi figure

    plt.plot(x_36,m1_36*(x_36-e_36)+f_36,color='red',label='y={}x+{}'.format(m1_36,-m1_36*e_36+f_36)) #Plot garis pertama
    plt.plot(x_36,m2_36*(x_36-i_36)+j_36,color='blue',label='y={}x+{}'.format(m2_36,-m2_36*i_36+j_36)) #Plot garis kedua
    plt.plot(tp_x_36,tp_y_36,color='green',marker='o',markersize=4) #Plot titik potong
    plt.legend(loc="best") #Menampilkan legend
# 6. Menampilkan hasil
    plt.show()
print('==========================================Terima Kasih==========================================')
```

```python
# 1. Import Library
import numpy as np
import matplotlib.pyplot as plt

print("--------------------------------------Yuk kita cari solusi dari bentuk persamaan garis--------------------------------------")
# 2. Definisikan Titik dan Penghitungan Gradien Garis Pertama
# Persamaan garis pertama melewati titik (a,b) dan (c,d)
a_36 = -1  # Titik (a,b)
b_36 = 0
c_36 = 2  # Titik (c,d)
d_36 = -2
m1_36 = (d_36 - b_36) / (c_36 - a_36)
print("Berikut merupakan gradien garis pertama: {:.2f}".format(m1_36)) #Mencetak gradien garis pertama

# 3. Menentukan Persamaan Garis Pertama
if -m1_36 * a_36 + b_36 > 0:
    print("Persamaan garis pertama adalah: y={:.2f}x+{:.2f}".format(m1_36, -m1_36 * a_36 + b_36)) #Mencetak persamaan garis pertama jika (-m1*a+ b>0)
else:
    print("Persamaan garis pertama adalah: y={:.2f}x{:.2f}".format(m1_36, -m1_36 * a_36 + b_36)) #Menctak persamaan garis pertama

# 4. Definisikan Titik dan Penghitungan Garis Sejajar
# Persamaan garis sejajar dengan garis pertama melalui titik (e,f)
e_36 = 5  # Titik (e,f)
f_36 = 3
m2_36 = m1_36  # Gradien yang sama karena sejajar
print("Berikut merupakan gradien garis sejajar: {:.2f}".format(m2_36)) #Mencetak gradien garis sejajar

if -m2_36 * e_36 + f_36 > 0:
    print("Persamaan garis sejajar adalah: y={:.2f}x+{:.2f}".format(m2_36, -m2_36 * e_36 + f_36)) #Mencetak persamaan garis sejajar jika (-m2*e+f>0)
else:
    print("Persamaan garis sejajar adalah: y={:.2f}x{:.2f}".format(m2_36, -m2_36 * e_36 + f_36)) #Mencetak persamaan garis sejajar

# 5. Plotting Garis-Garis
x_36 = np.linspace(-10, 10, 400)  # Membuat rentang nilai x yang luas

plt.figure(figsize=(8, 8), dpi=100)  # Menentukan ukuran dan resolusi figure

# 6. Plot garis pertama
if -m1_36 * a_36 + b_36 > 0:
    plt.plot(x_36, m1_36 * x_36 + (-m1_36 * a_36 + b_36), color='green', label='y={:.2f}x+{:.2f}'.format(m1_36, -m1_36 * a_36 + b_36))
else:
    plt.plot(x_36, m1_36 * x_36 + (-m1_36 * a_36 + b_36), color='green', label='y={:.2f}x{:.2f}'.format(m1_36, -m1_36 * a_36 + b_36))

# 7. Plot garis sejajar
if -m2_36 * e_36 + f_36 > 0:
    plt.plot(x_36, m2_36 * x_36 + (-m2_36 * e_36 + f_36), color='blue', label='y={:.2f}x+{:.2f}'.format(m2_36, -m2_36 * e_36 + f_36))
else:
    plt.plot(x_36, m2_36 * x_36 + (-m2_36 * e_36 + f_36), color='blue', label='y={:.2f}x{:.2f}'.format(m2_36, -m2_36 * e_36 + f_36))

plt.legend(loc='best')  #Menampilkan legend
plt.title('Plot Garis dan Garis Sejajar') #Menapilkan judul
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5) #Garis horizontal pada y=0
plt.axvline(0, color='black',linewidth=0.5) #Garis vertikal x=0

# 8. Mencetak Hasil
plt.show()

print("--------------------------------------Sampai bertemu lagi!--------------------------------------")
```

```python
# 1. Import Library
import numpy as np
import matplotlib.pyplot as plt

print("--------------------------------------Yuk kita cari solusi dari bentuk persamaan garis--------------------------------------")
# 2. Definisikan Titik dan Penghitungan Gradien Garis Pertama
# Persamaan garis pertama melewati titik (a,b) dan (c,d)
a_36 = -1  # Titik (a,b)
b_36 = 0
c_36 = 2  # Titik (c,d)
d_36 = -2
m1_36 = (d_36 - b_36) / (c_36 - a_36)
print("Berikut merupakan gradien garis pertama: {:.2f}".format(m1_36)) #Mencetak gradien garis pertama

# 3. Menentukan Persamaan Garis Pertama
if -m1_36 * a_36 + b_36 > 0:
    print("Persamaan garis pertama adalah: y={:.2f}x+{:.2f}".format(m1_36, -m1_36 * a_36 + b_36)) #Mencetak persamaan garis pertama jika (-m1*a+b>0)
else:
    print("Persamaan garis pertama adalah: y={:.2f}x{:.2f}".format(m1_36, -m1_36 * a_36 + b_36)) #Mencetak persamaan garis pertama

# 4. Definisikan Titik dan Penghitungan Gradien Garis Berhimpit
# Persamaan garis kedua berhimpit dengan garis pertama melalui titik (e,f) yang sama dengan titik (a,b)
e_36 = a_36  # Titik (e,f) yang sama dengan (a,b)
f_36 = b_36
g_36 = c_36  # Titik (g,h) yang sama dengan (c,d)
h_36 = d_36
m2_36 = m1_36  # Gradien yang sama karena berhimpit
print("Berikut merupakan gradien garis kedua: {:.2f}".format(m2_36)) #Mencetak gradien garis kedua

# 5. Menghitung Persamaan Garis Kedua Berhimpit
if -m2_36 * e_36 + f_36 > 0:
    print("Persamaan garis kedua adalah: y={:.2f}x+{:.2f}".format(m2_36, -m2_36 * e_36 + f_36)) #Mencetak persamaan garis kedua jika (-m2*e+f>0)
else:
    print("Persamaan garis kedua adalah: y={:.2f}x{:.2f}".format(m2_36, -m2_36 * e_36 + f_36)) #Mencetak persamaan garis kedua

# Karena garis-garis berhimpit, kita tahu bahwa gradien dan posisi mereka sama
print("Gradiennya sama dan garis-garis berhimpit, jadi ada tak hingga solusi.")

# 6. Plotting Garis-Garis
x_36 = np.linspace(-10, 10, 400)  # Membuat rentang nilai x yang luas

plt.figure(figsize=(8, 8), dpi=100)  # Menentukan ukuran dan resolusi figure

# 7. Plot garis pertama (dan kedua yang berhimpit)
if -m1_36 * a_36 + b_36 > 0:
    plt.plot(x_36, m1_36 * x_36 + (-m1_36 * a_36 + b_36), color='orange', label='y={:.2f}x+{:.2f}'.format(m1_36, -m1_36 * a_36 + b_36))
else:
    plt.plot(x_36, m1_36 * x_36 + (-m1_36 * a_36 + b_36), color='orange', label='y={:.2f}x{:.2f}'.format(m1_36, -m1_36 * a_36 + b_36))

# Menambahkan garis horizontal dan vertikal pada y=0 dan x=0
plt.axhline(0, color='black', linewidth=0.5)  #Garis horizontal pada y=0
plt.axvline(0, color='black', linewidth=0.5)  #Garis vertikal pada x=0

plt.legend(loc='best')  #Menampilkan legend
plt.title('Plot Garis yang Berhimpit') #Mencetak Judul
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# 8. Mencetak Hasil
plt.show()

print("--------------------------------------Sampai bertemu lagi!--------------------------------------")
```
